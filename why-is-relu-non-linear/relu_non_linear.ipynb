```python
import tensorflow as tf
import numpy as np
import pandas as pd
from pylab import rcParams
import matplotlib.pyplot as plt
import warnings
from mlxtend.plotting import plot_decision_regions
from matplotlib.colors import ListedColormap
```

```python
zero_one_colourmap = ListedColormap(('red', 'green'))
rcParams['figure.figsize'] = 14, 7
```

```python
df = pd.read_excel("Data.xlsx")
```

```python
X = df.loc[:, ["X1", "X2"]].values
y = df.loc[:, "Y"].values
```

```python
plt.scatter(X[:,0],X[:,1], c=y, s=100, cmap = zero_one_colourmap)
plt.show()
```

```python
def create_model(X, y, hidden_neurons, activation, total_epochs):
  model = tf.keras.models.Sequential()
  model.add(tf.keras.layers.Dense(hidden_neurons, 
                                  input_dim=2, 
                                  activation=activation))
  model.add(tf.keras.layers.Dense(1, activation=activation))
  model.compile(loss='binary_crossentropy',
                optimizer='adam',
                metrics=['binary_accuracy'])
  history = model.fit(X, y, epochs=total_epochs)
  return model
```

# Linear Model with 8 Neurons

```python
linear_model = create_model(X, y, 8, 'linear', 200)
```

```python
plot_decision_regions(X, y.ravel(), clf = linear_model, legend=2)
plt.show()
```

# Linear Model with 64 Neurons

```python
linear_model_64_neurons = create_model(X, y, 64, 'linear', 200)
```

```python
plot_decision_regions(X, y.ravel(), clf = linear_model_64_neurons, legend=2)
plt.show()
```

# Relu Activation with 8 neurons

```python
relu_8_neuron_model = create_model(X, y, 8, 'relu', 200)
```

```python
plot_decision_regions(X, y.ravel(), clf = relu_8_neuron_model, legend=2)
plt.show()
```

# Relu Activation with 64 neurons

```python
relu_64_neuron_model = create_model(X, y, 64, 'relu', 200)
```

```python
plot_decision_regions(X, y.ravel(), clf = relu_64_neuron_model, legend=2)
plt.show()
```
